### Task 1.3

````python
result = []
pts = []

for index, row in data.iterrows():
    if row['GF'] > row['GA']:
        result.append('W')
        pts.append(3)
    elif row['GF'] == row['GA']:
        result.append('D')
        pts.append(1)
    else:
        result.append('L')
        pts.append(0)

data['result'] = result
data['pts'] = pts
data
````

### Task 1.4

````python
for x, y in data.groupby('Opponent'):
    count = sum(y.pts)
    print(f"Play Againest {x}: {count}")
````

### Task 1.5

````python
count = 0
cumu = []
for _, r in data.iterrows():
    count += r.pts
    cumu.append(count)
data['cumu'] = cumu
data
````

````python
cumu2 = data.pts.cumsum()
safe = np.where(cumu2 >= 40)
safe
````
